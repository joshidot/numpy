## Fourier Series

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Basis functions ($\sin(\theta), \sin(2\theta) ...$)

```python
fs = 1000
theta = np.arange(0,2*np.pi,2*np.pi/fs)

plt.figure(figsize = [20,6])

for i in range(1,6):
    amp = 1*1/float(i)
    freq = 0.025*i
    
    '''Y= A sin(n.theta)'''
    y = amp*np.sin(i*theta)
    
    plt.plot(theta,y, label= "sin("+str(i)+"theta)")
    
plt.xlabel("Angular frequency in PI") 
plt.ylabel("Displacement")
plt.legend()
plt.show()
```

|  Basis function |  Periodicity |  
|---|---|---|---|---|
| $sin(\theta)$ |  2pi |   
|$sin(2\theta)$  | pi  |  
| $sin(3\theta)$  | 2pi/3  |  
| $sin(4\theta)$  | pi/4 |

### Fourier Series: [Square Wave](https://mathworld.wolfram.com/FourierSeriesSquareWave.html)

![img](https://mathworld.wolfram.com/images/equations/FourierSeriesSquareWave/NumberedEquation3.gif)

#### Individual basis

```python
L = 1
xs =1000
x = np.arange(0,2,10/xs)

plt.figure(figsize = [15,4])

for n in [i for i in range(10) if i%2 !=0]:
    '''creating individual terms in the series'''
    fi = (4/np.pi)*(1/n)*np.sin(n*np.pi*x/float(L))
    plt.plot(x,fi)
```

### Subsequent development

```python
L = 1
xs =1000
x = np.arange(0,2,10/xs)
plt.figure(figsize = [15,4])

for n in [i for i in range(10) if i%2 !=0]:
    if n == 1:
        '''taking care of first term'''
        fi = (4/np.pi)*(1/n)*np.sin(n*np.pi*x/float(L))
        plt.plot(x,fi, label = "n="+str(n))
        ff = fi
    else:
        '''terms following first term'''
        fi = ff + (4/np.pi)*(1/n)*np.sin(n*np.pi*x/float(L))
        plt.plot(x,fi, label = "n="+str(n))
        ff = fi
plt.legend()
plt.xlabel("x")
plt.ylabel("fx")
plt.show()
```

#### Sum of all n-terms

```python
L = 1
xs =1000
x = np.arange(0,10,10/xs)
plt.figure(figsize = [15,4])

'''including n = 1,3,5,...19th terms in basis expansison'''
fx = sum([(4/np.pi*(1/n)*np.sin(n*np.pi*x/float(L))) for n in [i for i in range(20) if i%2 !=0]])

plt.plot(x,fx,label ="sawtooth wave")
plt.legend()
plt.xlabel("x")
plt.ylabel("fx")
plt.show()
```
